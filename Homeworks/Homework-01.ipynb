## Problem 1:
If we want to write a program checking whether a positive integer $p$ is in the form of 
$$p = F_n = 2^{n} + 1, \quad \text{ where } n \in \mathbb{Z}, n>0.$$
We basically can check whether there exists an integer $n$ such that the relation above is satisfied.

```python
def isMyinteger(p):
    if type(p) == int:
        n = 0
        while 2**n + 1 <= p:    # while loop will be featured in Lecture 5
            if 2**n + 1 == p:
                return True
            n = n + 1
        return False
    else: 
        print("The input is not an integer.")
```

```python
isMyinteger(9) # this should be True
```

```python
isMyinteger(8) # this should be False
```

```python
isMyinteger(-3.14) # this should print an error
```

## Problem 3
How do we graph a function $f(x,y)$ in two variables? One way to do it is to consider the function's value $f(x,y)$ at a point $(x,y)$ as the color-intensity at that point, in `matplotlib`, this is called "grayscale".

For example, if $f(0,0) = 0$, then we put a black pixel at $(0,0)$, i.e. the center. 

If $f(1, 1) = 1$, then we put a white pixel at $(1,1)$, i.e. the top-right. 

If $f(x_1,y_1)= 0.5$, then we use a gray pixel at point $(x_1,y_1)$.

For simplicity, the grayscale is from 0 to 1, in that,
when $f(x,y) > 1$, we truncated these values to be 1 and use white pixels to represent these values;
when $f(x,y) < 0$, we truncated these values to be 0 and use black pixels to represent these values.

Above procedure is an example of "normalization" commonly used in machine learning.
